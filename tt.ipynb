```python
import requests
from bs4 import BeautifulSoup
r = requests.get('https://kemono.party/fanbox/user/50010871')
print(r.text)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
test = soup.find_all('h2', class_='post-card__heading')
#print(test)
for i in test:
    print(i.find("a").get("href"))
    print(i.find("a").get_text())
```

```python
r2 = requests.get('https://kemono.party/fanbox/user/4588267/post/4243218')
print(r2.text)
```
